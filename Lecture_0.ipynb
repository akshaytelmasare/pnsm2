# Tossing A Coin

```python
def coin_toss():
    x = rd.random()
    if x > 0.5:
        return "Head"
    else:
        return "Tail"
    
def number_of_heads(k):
    heads_count = 0
    for i in range(k):
        coin_toss_result = coin_toss()
        if coin_toss_result == "Head":
            heads_count += 1
    return heads_count
```

```python
k = 100
h = number_of_heads(k)

print(h)
```

```python
n = 10000
k = 100
arr = []
for i in range(n):
    h = number_of_heads(k)
    arr.append(h)

plt.hist(arr, bins= k+1);
```

```python
dice = [1,2,3,4,5,6]
def dice_roll():
    return rd.choice(dice)

def sum_n_dice_roll(k):
    sum = 0
    for i in range(k):
        sum += dice_roll()
    return sum
```

```python
accumated = []
n = 10_000
k = 120
for i in range(n):
    sum_k = sum_n_dice_roll(k)
    accumated.append(sum_k)

plt.hist(accumated, bins=4*k);
```

```python
accumated = []
n = 10_000
k = 100
for j in range(n):
    z = [rd.random() for i in range(k)]
    sum_z = sum(z)
    accumated.append(sum_z)

plt.hist(accumated, bins=100);
```
